```python
# 2.a
# question 1) |sin(πx)| = ln(x)
# 2) |sin(πx)| = |ln(x)|

x2 = np.arange(0.01,4, 0.1)
y1 = np.log(x2)
y1_1 = abs(np.log(abs(x2)))
y2 = abs(np.sin(np.pi*x2))
```

```python
# change y1 -> y2 to see changes
plt.plot(x2, y2, label="|sin(πx)|", color="green") 
plt.plot(x2, y1_1, label="ln(x)", color="blue")

int_indices = np.argwhere(np.isclose(y1_1, y2, atol=0.07)).flatten()
x_int = x2[int_indices] 
y_int = y2[int_indices] 

plt.scatter(x_int, y_int , color="red", zorder= 5, label="Intersection")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
# number of solution = 4
#  note i had to try different vals of absolute tolerance to get the desired points
```

```python
x_n = np.arange(0, 2*np.pi, 0.1)
m = np.sqrt(3)/2
n = np.cos(x_n)
o = np.sin(x_n)
no_max = np.max([n, o])
```

```python
int_vals_n = np.argwhere(np.isclose(n, m ,atol=0.03)).flatten()
int_vals_o = np.argwhere(np.isclose(o, m ,atol=0.025)).flatten()

combined_indices = np.unique(np.concatenate((int_vals_n, int_vals_o)))

x_int_n = x_n[int_vals_n]
y_int_n = n[int_vals_n]
x_int_o = x_n[int_vals_o]
y_int_o = o[int_vals_o]

horizontal_line_y = np.sqrt(3)/2
# plt.yticks([0, 1, horizontal_line_y], ['0', '1', r'$\frac{1}{8}$'])

plt.scatter(x_int_n, y_int_n, color="red",zorder=5, label="intersection" )
plt.scatter(x_int_o, y_int_o, color="red",zorder=5)

plt.xlabel("x")
plt.ylabel("y")
plt.axhline(y=m, label="√3/2", color='black', linestyle='-', linewidth=2)
plt.plot(x_n, n, label="cos(x) - n", color="green")
plt.plot(x_n, o, label="sin(x) - o", color="blue")
plt.legend()
```

```python
# 2sin(|x|) = (|x| - 1)/2
x_lin = np.linspace(-2 * np.pi, 2 * np.pi, 200)
p = np.sin(abs(x_lin))
q = (abs(x_lin) - 1)/2
```

```python
inter_p = np.argwhere(np.isclose(q, p ,atol=0.03)).flatten()
x_int_p = x_lin[inter_p]
y_int_p = p[inter_p]
```

```python
plt.scatter(x_int_p, y_int_p, color="red",zorder=5, label="intersection" )
plt.plot(x_lin, q, color="green", label="(|x|-1)//2")
plt.plot(x_lin, p, color="blue", label="sin|x|")
plt.grid()
plt.legend()
plt.show()
```

```python
x_vec = np.linspace(0, 3*np.pi, 200)
r = 3* abs(np.cos(x_vec))
s = 5* abs(np.sin(x_vec))
```

```python
intersection = np.argwhere(np.isclose(s, r ,atol=0.15)).flatten()
x_rs = x_vec[intersection]
y_rs = s[intersection]
```

```python
plt.scatter(x_rs, y_rs, label="intersection", color='red', zorder=5)
plt.plot(x_vec, r, color="green", label="3|cosx|")
plt.plot(x_vec, s, color="blue", label="5|sinx|")
plt.grid()
plt.legend()
```
